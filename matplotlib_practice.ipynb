```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Create a Line Chart using the MATLAB Method

```python
# create the plot.
plt.plot(x_axis, y_axis)
```

# Create a Line Chart using the Object-Oriented Interface Method

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

### Another way to create a line chart using the object-oriented interface method:

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

### Create a quick graph using object-oriented interface method:

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

### To change figure-level attributes (e.g., axis labels, a title, or a legend) or save the figure as an image:

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = "Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare By Month")
# Add legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

# Create the plot using object-oriented method

```python
# change the line color to green
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# create label for x axis and y axis
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# set the y limit
ax.set_ylim(0, 45)
# set the title
ax.set_title("PyBer Fare By Month")
# Add a grid
ax.grid()
ax.legend()
```

# Create Bar Charts Using the MATLAB Approach

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

### Create a Vertical Bar Chart

```python
# create the plot.
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
```

### NOTE: To invert the y-axis to have months in ascending order, use the 'gca()' method which means 'get current axes'and chain it to .invert_yaxis() method.

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

## Create the plot using matplotlib MATLAB plotting method

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

 NOTE: We can refactor our code so that we don't use a for loop to create the new list (y_axis_larger) and use Python's list comprehension instead. You can use a list comprehension to replace many for and while loops. The format for list comprehension: new_list = [expression for item in list if conditional]

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

Using the object-oriented approach, make the following changes to your bubble chart:

1. Change the color of the markers to sky blue.

2. Change the size of the markers to 5 times each data point.

3. Make the color 20% transparent.

4. Add a black edge color to the circles.

5. Make the linewidth of the circles 2 points.

6. Add a legend for the city of Boston.

7. Add a title.

8. Switch the axis so that the Fare($) data is on the x-axis.

9. Add a limit for the x-axis data.

10. Invert the y-axis so that January is at the top.

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, 
           x_axis, 
           fc='#75bbfd', 
           alpha= 0.5, 
           edgecolor="black", 
           linewidth=2, 
           s = [i * 5 for i in y_axis], 
           label="Boston")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.legend()
```

To fix the crowding of the pie chart, increase the size of the pie chart using the first line of code before the rest.

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

Using the object-oriented approach, make the following changes to the pie chart:

1. Add a percentage to one decimal place to each wedge of the pie.

2. Increase the figure size to 8x8.

3. Explode the two highest percentage months.

4. Add a shadow.

5. Add a start angle so that January is at the top.

6. Reverse the order so that the month order is in a clockwise direction.

7. Add new colors of your own choosing or use the colors from the previous pie chart.



```python
fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", 
          "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       shadow=True,
       startangle=90,
       counterclock=False,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```
